```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
sns.lmplot(x="AT", y="PE", data=df, scatter_kws={"s": 10, "alpha":0.5}, line_kws={"color": "red"})
```
