# Exploring statistical values of data

#### Importing Libraries

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

#### Reading data

```python
df = pd.read_csv('Data.csv')
df.head()
```

#### Extracting features and label

```python
X = df[['AT', 'V', 'AP', 'RH']]
y = df['PE']

X = sm.add_constant(X)
```

#### Evaluating Ordinary Least Squares (OLS) model and displying summary of relevant results

```python
model = sm.OLS(y, X).fit()
print(model.summary())
```
