# Pandas SF Salaries Exercise 

We are using the [SF Salaries Dataset](https://www.kaggle.com/kaggle/sf-salaries) from Kaggle!

#### Import pandas and NumPy

```python
import pandas as pd
import numpy as np
```

#### Read Salaries.csv as a dataframe called sal

```python
sal = pd.read_csv("Salaries.csv")
```

#### Checking the head and tail of the DataFrame

```python
sal.head()
```

```python
sal.tail()
```

#### Using the .info() method to find out how many entries there are in each column

```python
sal.info()
```

#### Deleting column with no values ('Notes')

```python
sal = sal.drop('Notes', axis=1)
```

#### Replacing 'Not Provided' values with np.nan (in numeric columns)

```python
num_columns = ['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits']
sal[num_columns] = sal[num_columns].replace('Not Provided', np.nan)
```

#### Changing type of numeric columns to float

```python
sal[num_columns] = sal[num_columns].astype(float)
```

```python
sal.info()
```

#### Get the average of a BasePay

```python
sal['BasePay'].mean()
```

#### Get the highest amount of OvertimePay

```python
sal["OvertimePay"].max()
```

#### How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?)

```python
sal[sal['Year'] == 2013]['JobTitle'].value_counts() == 1
```

```python
sum(sal[sal['Year'] == 2013]['JobTitle'].value_counts() == 1)
```

#### How many people have the word Chief in their job title?

```python
def is_chief(title):
    return "chief" in title.lower()
```

```python
sal["JobTitle"]
```

```python
sal["JobTitle"].apply(lambda x: is_chief(x))
```

```python
sum(sal["JobTitle"].apply(lambda x: is_chief(x)))
```

### Explore with Seaborn

#### Import Seaborn and Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

#### Show Histogram of entire BasePay and by Year

```python
sns.histplot(data=sal, x="BasePay", bins=20)
plt.show()
```

```python
sns.histplot(data=sal, x='BasePay', hue='Year', multiple="stack")
plt.show()
```

```python
sns.displot(data=sal, x='BasePay', col='Year', kind='hist')
plt.show()
```

#### Show Joint Plot of Base Pay and Total Pay

```python
sns.jointplot(data=sal, x='BasePay', y='TotalPay')
plt.show()
```
