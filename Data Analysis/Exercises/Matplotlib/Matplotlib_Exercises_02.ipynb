#### Importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Read data with Python built-in functions

```python
def getUSPop(fileName):
    with open(fileName, 'r') as f:
        data = f.read()
    lines = data.splitlines()
    dates, pops = [], []
    for line in lines:
        date, pop = line.split()
        dates.append(int(date))
        for char in pop:
            if char == ',':
                pop = pop.replace(char, '')
        pops.append(int(pop))
    return dates, pops
```

#### Read the data using Pandas

```python
import pandas as pd
data = pd.read_csv('USPopulation.txt', sep=' ', thousands=',')
```

#### Regular plot

```python
# dates, pops = data.iloc[:, 0], data.iloc[:, 1]
dates, pops = getUSPop('USPopulation.txt')
plt.plot(dates, pops)
plt.title('US Population Growth')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

#### Semi log y plot

```python
plt.figure(figsize=(10, 6))
plt.semilogy(dates, pops)
plt.title('US Population Growth')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

#### Read data with Python built-in functions

```python
def getCountryPops(fileName):
    with open(fileName, 'r') as f:
        data = f.read()
    lines = data.splitlines()
    pops = {}
    for line in lines:
        items = line.split('\t')
        country = items[1]
        pop = items[2]
        for char in pop:
            if char == ',':
                pop = pop.replace(char, '')
        pops[country] = int(pop)
    return pops
```

#### Read the data using Pandas

```python
data = pd.read_csv('CountryPops.txt', sep='\t', thousands=',')
```

```python
# countries = data.iloc[:, 1]; pops = data.iloc[:, 2]
pops = getCountryPops('CountryPops.txt')
countries = list(pops.keys())
populations = list(pops.values())
plt.plot(populations)
plt.title('Country Populations')
plt.xlabel('Country Rank Based on Size')
plt.ylabel('Population')
plt.show()
```

#### Semi log y plot

```python
plt.semilogy(populations)
plt.title('Country Populations')
plt.xlabel('Country Rank Based on Size')
plt.ylabel('Population')
plt.show()
```

#### Investigate first digits of populations to show Benford's law

```python
firstDigits = [int(str(pop)[0]) for pop in populations]
plt.hist(firstDigits, bins=np.arange(1, 11) - 0.5, density=True)
plt.title('Distribution of First Digits in Country Populations')
plt.xlabel('First Digit')
plt.ylabel('Density')
plt.show()
```
